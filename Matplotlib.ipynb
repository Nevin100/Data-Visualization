# Pyplot :

```python
# x-axis :
x = ['Monday','Tuesday','Wednesday','Thursday','Friday']

# y-axis :
y = [10,20,40,60,5]

#title:
plt.title("Bakery Sales :")

#xlabel
plt.xlabel("Days :")

#ylabel
plt.ylabel("Days :")

plt.plot(x,y)
plt.show()
```

# Pyplot Fuctions :


```python
# 1.) plt.plot(x,y, {rest optional...}) --> plottig purpose 
# 2.) plt.title() --> Title Heading of the Plot
# 3.) plt.xlabel() & plt.ylabel() --> labels for x and y
# 4.) plt.show() --> To Show the plot
# 5.) plt.legend() --> To Show the meaning label of line
# 6.) plt.grid() --> to show the grid behind the plot
# 7.) plt.xlim() --> to set the x limit of x axis
# 8.) plt.ylim() --> to set the y limit of y axis

months = ['January','February', 'March', 'April' ]
sales = [5000,4000,6600,2000]

plt.title("Sales Graph -->")

plt.xlabel("Months Q1-->")
plt.ylabel("Sales in $-->")

plt.plot(months, sales, color='blue', linestyle='--', linewidth=1, marker='o', label='2025 Sales Data')
plt.legend()
plt.grid(color='gray')
plt.ylim(0, 10000)
plt.show()
```

# Charts :

```python
# Bar charts : --> Category Comparison and Their Purposes
# plt.bar(x, height, color='', width=value, label='label name')

products = ['A', 'B', 'C', 'D']
sales = [1000, 2000, 5000, 3000]

plt.bar(products, sales, color='yellow', label='2000 Sales Data')
plt.title("Bar Chart ")
plt.xlabel("Products --> ")
plt.ylabel("Sales --> ")
plt.legend()
plt.plot()
plt.show()
```

```python
# pie Charts --> Proportion illustration and Whole Representation

# plt.pie(values, labels=label_list, colors=color_list, autopct='%1.1f%%)

regions = ['North','South','West','East']
revenue = [3000, 2000, 15000, 10000]

plt.pie(revenue, labels=regions, colors=['gold', 'silver','pink', 'coral'], autopct='%1.1f%%')
plt.title("Revenue Pie Chart :")
plt.plot()
plt.show()
```

```python
# Histogram--> Numerical Distubution and Insights
# plt.hist(data, bins=number_of_bins, color='color_name', edgecolor='black')

scores =[45,67,78,45,96,85,74,25,14,52,63,30,2,14,65,84,56,97]
plt.hist(scores, bins=3, color='coral', edgecolor='black')
plt.title("Histogram Of Marks ")
plt.ylabel("Number of Student --> ")
plt.xlabel("Range of Marks --> ")
plt.plot()
plt.show()
```

# Scatter Plots :

```python
# Scatter Plot : --> Pattern Insughts and Correlation
# plt.scatter(x, y, color='color_name', marker='marker_style', label='label_name')

hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_marks= [50, 55, 60, 65, 70, 75, 80, 85, 90]

plt.scatter(hours, exam_marks, color='green', marker='o', label='Student Marks')

plt.xlabel("Hours -->")
plt.ylabel("Exam Marks -->")
plt.title("Scatter plot")
plt.grid(linestyle='--')
plt.legend()
plt.plot()
plt.show()
```

```python
hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_marks1= [50, 55, 60, 65, 70, 75, 80, 85, 90]
exam_marks2=[20,30,40,50,60,70,80,90,10]

plt.scatter(hours, exam_marks1, color='coral', marker='o', label='Student Marks Class A')
plt.scatter(hours, exam_marks2, color='blue', marker='^', label='Student Marks Class B')

plt.xlabel("Hours -->")
plt.ylabel("Exam Marks -->")
plt.title("Scatter plot")
plt.grid(linestyle='--')
plt.legend()
plt.plot()
plt.show()
```

# Sub Plots and Functions :

```python
# plt.subplot() --> To create multiple plots in a single figure
# plt.subplot(nrows, ncols, index)

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.subplot(1,2,1)  # 1 row, 2 columns, first subplot
plt.plot(x, y, color='red', label='Line Chart')
plt.title("First Subplot")

plt.subplot(1,2,2)  # 1 row, 2 columns, second subplot
plt.bar(x, y, color='blue', label='Bar Chart')
plt.title("Second Subplot")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.plot()
plt.show()
```

```python
# More Control and Recommended : 

#fig, ax = plt.subplots(figsize=(10, 5))  # Create a figure and axis with specified size

fig, ax = plt.subplots(1, 2, figsize=(12, 6))  # Create a figure with 1 row and 2 columns

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]

ax[0].plot(x, y1, color='green', marker='o', label='Line Chart')
ax[0].set_title("Line Chart")

y2 = [5, 15, 25, 35, 45]
ax[1].bar(x, y2, color='orange', label='Bar Chart')
ax[1].set_title("Bar Chart")

plt.suptitle("Combined Plots")  # Overall title for the figure
ax[0].set_xlabel("X-axis")
ax[0].set_ylabel("Y-axis")

ax[1].set_xlabel("X-axis")
ax[1].set_ylabel("Y-axis")
ax[0].legend()
ax[1].legend()
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the combined plots
```
