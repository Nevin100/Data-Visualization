```python
# x-axis :
x = ['Monday','Tuesday','Wednesday','Thursday','Friday']

# y-axis :
y = [10,20,40,60,5]

#title:
plt.title("Bakery Sales :")

#xlabel
plt.xlabel("Days :")

#ylabel
plt.ylabel("Days :")

plt.plot(x,y)
plt.show()
```
