# Pyplot :

```python
# x-axis :
x = ['Monday','Tuesday','Wednesday','Thursday','Friday']

# y-axis :
y = [10,20,40,60,5]

#title:
plt.title("Bakery Sales :")

#xlabel
plt.xlabel("Days :")

#ylabel
plt.ylabel("Days :")

plt.plot(x,y)
plt.show()
```

# Pyplot Fuctions :


```python
# 1.) plt.plot(x,y, {rest optional...}) --> plottig purpose 
# 2.) plt.title() --> Title Heading of the Plot
# 3.) plt.xlabel() & plt.ylabel() --> labels for x and y
# 4.) plt.show() --> To Show the plot

months = ['January','February', 'March', 'April' ]
sales = [5000,4000,6600,2000]

plt.title("Sales Graph -->")

plt.xlabel("Months Q1-->")
plt.ylabel("Sales in $-->")

plt.plot(months, sales, color='blue', linestyle='--', linewidth=2, marker='o')
plt.show()
```
